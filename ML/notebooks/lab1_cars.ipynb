### Заполнение пропусков данных

```python
mean_engcap = data['engine_capacity'].mean()  
data['engine_capacity'] = data['engine_capacity'].fillna(mean_engcap)
print(data.isnull().sum()) # Проверка 
```

### Удаление дубликатов

```python
data.duplicated().sum()
data = data.drop_duplicates()
data.duplicated().sum()
```

### Обогащение данных

```python
# Возраст авто
data['car_age'] = 2025 - data['year_produced']
print(data[['year_produced', 'car_age']].head())
```

```python
data = data.drop(columns=['year_produced'])
```

```python
# Средний пробег в год
data['mileage'] = data['odometer_value'] / data['car_age']
print(data[['odometer_value', 'car_age', 'mileage']].head())
```

### Преобразование данных в категории

```python
categories = ['manufacturer_name', 'model_name', 'transmission', 'color', 'engine_fuel',
              'engine_type', 'body_type', 'state', 'drivetrain', 'location_region']
for i in categories:
    data[i] = data[i].astype('category')
```

### Количество уникальных значений в категории

```python
for col in categories:
    print(f"{col}: {data[col].nunique()} unique values")
```

### Обобщение очень редких категорий в Other

```python
for col in categories:
    # Находим категории, которые встречаются менее чем в 0,005% строк
    threshold = 0.005 * len(data)
    rare_categories = data[col].value_counts()[data[col].value_counts() < threshold].index
    
    # Добавляем категорию 'Other', если её ещё нет
    if 'Other' not in data[col].cat.categories:
        data[col] = data[col].cat.add_categories('Other')
    
    # Заменяем редкие категории на 'Other'
    data[col] = data[col].replace(rare_categories, 'Other')
    
    # Удаляем неиспользуемые категории (опционально)
    data[col] = data[col].cat.remove_unused_categories()
```

### Обогащение данных 2

```python
# Популярность модели
model_counts = data['model_name'].value_counts()

# Присвоение порядкового номера по популярности
popularity_rank = model_counts.rank(method='min', ascending=False).astype(int)

# Создание столбца и вывод
data['popularity_rank'] = data['model_name'].map(popularity_rank)
print(data[['model_name', 'popularity_rank']].head())
```

```python
# Ранг производителя
manuf_counts = data['manufacturer_name'].value_counts()

# Присвоение порядкового номера по рангу
manuf_rank = manuf_counts.rank(method='min', ascending=False).astype(int)

# Создание столбца и вывод
data['manuf_rank'] = data['manufacturer_name'].map(manuf_rank)
print(data[['manufacturer_name', 'manuf_rank']].head(1000))
```

### Исключение выбросов

```python
outlier = data[columns]
outlier
```

```python
Q1 = outlier.quantile(0.25)
Q1
```

```python
Q3 = outlier.quantile(0.75)
Q3
```

```python
IQR = Q3-Q1
data_filtered = outlier[~((outlier < (Q1 - 1.5 * IQR)) |(outlier > (Q3 + 1.5 * IQR))).any(axis=1)]
data_filtered
```

```python
index_list = list(data_filtered.index.values)
```

```python
# Хранить данные без выбросов в отдельном DataFrame.
data_filtered = data[data.index.isin(index_list)]
data_filtered.head()
```

```python
filtered_outlier = data_filtered[columns]

plt.figure(figsize=(15, 10))

for i, column in enumerate(filtered_outlier, 1):
    plt.subplot(4, 3, i) 
    data[column].hist(log=True, bins=50) 
    plt.title(column)
    plt.grid(True, which="both", ls="-", alpha=0.3) 

plt.tight_layout()  
plt.show()
```
